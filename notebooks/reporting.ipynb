```python
import sys
sys.path.append("../src")
import pandas as pd
from pandas import Series
from utils.models import get_models, Model
```

```python
# get all 

# json_valid
# model, zero_shot, few_shot, cot, self_consistency

# EM
# model, zero_shot, few_shot, cot, self_consistency

# LLM_as_judge
# model, zero_shot, few_shot, cot, self_consistency

# latency
# model, zero_shot, few_shot, cot, self_consistency

models = get_models()
```

```python
def get_series(model_name, file, name):
    df = pd.read_csv(f"../results/{model_name}/{file}")
    return df[name]
```

```python
def calc_mean_json_valid(is_valid: Series) -> float:
    return is_valid.mean()
```

```python
def calc_mean_accuracy(score: Series) -> float:
    return score.mean()
```

```python
def report(metric, models: list[Model], calc):
    rows = []
    for model in models:
        rows.append({
            "model": model.short_name,
            "zero_shot": calc(get_series(model.short_name, "zero_shot_prompt.csv", metric)),
            "few_shot": calc(get_series(model.short_name, "few_shot_prompt.csv", metric)),
            "cot": calc(get_series(model.short_name, "cot_prompt.csv", metric)), 
            # "self_consistency": calc(get_series(model.short_name, "self_consistency_prompt.csv", metric)),
        })
    df = pd.DataFrame(rows)
    df.to_markdown(f"../results/reports/{metric}.md", index=False)
```

```python
# json_valid
report("is_valid_json", models, calc_mean_json_valid)
```

```python
# accuracy
report("score", models, calc_mean_accuracy)
```

```python
# latency
report("latency", models, lambda a: a.mean())
```
