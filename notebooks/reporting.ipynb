```python
import sys
sys.path.append("../src")
import pandas as pd
from pandas import Series
from utils.models import get_models, Model
```

```python

models = get_models()
```

```python
def get_series(model_name, file, name):
    df = pd.read_csv(f"../results/{model_name}/{file}")
    return df[name]
```

```python
def report(metric, models: list[Model], calc, file_name):
    rows = []
    for model in models:
        rows.append({
            "model": model.short_name,
            "zero_shot": calc(get_series(model.short_name, "zero_shot_prompt.csv", metric)),
            "few_shot": calc(get_series(model.short_name, "few_shot_prompt.csv", metric)),
            "cot": calc(get_series(model.short_name, "cot_prompt.csv", metric)), 
            "self_consistency": calc(get_series(model.short_name, "self_consistency_prompt.csv", metric)),
            "pal": calc(get_series(model.short_name, "pal_prompt.csv", metric)),
        })
    df = pd.DataFrame(rows)
    df.to_markdown(f"../results/reports/{file_name}.md", index=False)
```

```python
# json_valid
report("is_valid_json", models, lambda a: a.mean(), file_name="is_valid_json")
```

```python
# accuracy
report("score", models, lambda a: a.mean(), file_name="tolerant_accuracy")
```

```python
# mre
report("re", models, lambda a: a.mean(), file_name="mre")
```

```python
# latency
report("latency", models, lambda a: a.mean(), file_name="latency")
```
